```python
%matplotlib inline
import pandas as pd
import os
import pylab as plt
import numpy as np
import matplotlib.cm as cm
from sklearn import ensemble, cross_validation, preprocessing, grid_search 
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, explained_variance_score

plt.rcParams['figure.figsize'] = (12.0, 8.0) # Set the plot size
```

```python
# Read data
data_path = "../ios_gb_IPHONE_GAMES_01-10-2014_01-12-2014_freeDownloads/"
files = os.listdir(data_path)
for i, f in enumerate(files):
  temp = pd.read_csv(open(data_path + f, 'r'))
  if i ==0:
     data = pd.DataFrame(columns=temp.columns)
  data = data.append(temp[pd.notnull(temp['outputActualDownloads'])]) 

# Convert Dates to useful features
data['Date'] = pd.to_datetime(data['Date'])
data['weekday'] = [d.weekday() for d in data['Date']]
data['month'] = [d.month for d in data['Date']]
data['yday'] = [d.timetuple().tm_yday for d in data['Date']]

# Cleanup
del data['Date']
del data['IAP'] # All = 1 so not useful
```

Let's have a look at the data set probability distribution of all the features (on the diagonal) as well as the relations between them.

```python
junk = pd.scatter_matrix(data, diagonal='kde')
```

Now let's look at the app distribution

```python
names = data['nameFreeApplication'].unique()
data_per_app = [len(data['nameFreeApplication'][data['nameFreeApplication'] == n]) for n in names]
print "How many data points per app: ", data_per_app
print "Average data points per app: ", np.mean(data_per_app)
```

```python
colors = cm.gist_ncar(np.linspace(0, 1, len(names) + 1))
for i, n in enumerate(names):
  indexes = data['nameFreeApplication'] == n
  plt.scatter(data['ranks'][indexes], data['outputActualDownloads'][indexes], color=colors[i], label=n)
plt.xlabel('Rank')
plt.ylabel('Downloads')
plt.legend()  
plt.show()
```

Seems like there is a slight bias in Dowloads per app. See the writeup for comments.

Now let's build our Random Forest model.

```python
# Cleanup
y_all = data['outputActualDownloads'].values
del data['outputActualDownloads']
del data['nameFreeApplication']

print "Model features: ", data.columns
x_all = data.values

# Split test set - keep 25% of the dataset at random for the test set
x_train, x_test, y_train, y_test = cross_validation.train_test_split(x_all, y_all, test_size=0.25, random_state = 23)

# Grid search - Search through a list of parameters to find the best model
clf = ensemble.RandomForestRegressor()
param_grid = [
  # RF
  {'n_estimators':[300, 400, 450], 'max_depth':[None]},
 ]
clfg = grid_search.GridSearchCV(clf, param_grid, cv=3, verbose=0, n_jobs=1, refit=True)

scores = cross_validation.cross_val_score(clfg, x_train, y_train, cv=3, n_jobs=1)
print "CV training results: ", scores
print("Mean R2: %0.2f (+/- %0.2f)\n" % (scores.mean(), scores.std() * 2))   
print "With model: ", clfg.estimator
```

```python
# Feature importance
clfg.fit(x_train, y_train)
bf = dict(zip(clfg.best_estimator_.feature_importances_, data.columns.values))
s = sorted(bf.keys())[::-1]
print "Features importance:"
for ss in s:
    print '%.2f' % (ss * 100), "% " + bf[ss]
```

As expected the most important feature is the ranks, then day-to-day variability comes next. The weekly relationship seems quite weak and we clearly lack data to see any monthly relationship. 

Now let's **test the model** on the unseen test set to make sure it generalize well.

```python
# Test
y_pred_test = clfg.predict(x_test)
mae = mean_absolute_error(y_test, y_pred_test)
mse = mean_squared_error(y_test, y_pred_test)
r2 = r2_score(y_test, y_pred_test) 
ev = explained_variance_score(y_test, y_pred_test)
print "Test set results"
print "MAE: ", mae
print "MSE: ", mse
print "R2: ", r2
print "EV: ", ev
```

```python
plt.scatter(x_test[:,0], y_test, color='blue', label='True value')
plt.scatter(x_test[:,0], y_pred_test, color='red', label='Prediction')
plt.title("Test data results")
plt.xlabel("Rank")
plt.ylabel("Downloads")
plt.legend()
plt.show()
```

The predictions seems very reasonable.

```python
error = y_test - y_pred_test
plt.hist(error, bins=100, alpha=0.5)
plt.xlabel('Error [y_test - y_pred_test]')
plt.ylabel('N')
plt.title('Error distribution')
plt.show()
```

The error is roughly normaly distributed, so everything looks good.
